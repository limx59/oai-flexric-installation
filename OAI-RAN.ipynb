# Tutorial: Instalação da RAN (gNodeB) e FlexRIC

 
# Passo 1: Instalar dependências

Primeiramente, deve-se instalar as dependências. Em um terminal, execute os seguintes comandos:

```bash
sudo add-apt-repository ppa:ettusresearch/uhd

sudo apt-get install libuhd-dev uhd-host -y

sudo apt-get install libuhd4.2.0 -y

sudo apt-get install libuhd4.4.0 -y

sudo apt-get install -y automake 

sudo apt-get install -y g++ make libpcre2-dev byacc 

sudo apt-get install -y cmake python3-dev libsctp-dev
```

# Passo 2: Resolver problema de versão da biblioteca libuhd

Ao executar os comandos abaixo, caso apareça o erro "cannot access archive", pode seguir para o próximo passo

```bash
sudo dpkg -i --force-overwrite /var/cache/apt/archives/libuhd4.2.0_4.2.0.1-0ubuntu1~focal1_amd64.deb

sudo dpkg -i --force-overwrite /var/cache/apt/archives/libuhd4.4.0_4.4.0.0-0ubuntu1~focal1_amd64.deb
```

# Passo 3: Instalar o software SWIG

O SWIG é um software que serve para conectar programas escritos em C/C++ com programas escritos em diversas linguagens de alto nível. Para instalá-lo, execute os comandos abaixo.

```bash
git clone https://github.com/swig/swig.git

cd PATH_TO/swig

./autogen.sh

./configure --prefix=/usr/

make

sudo make install
```

# Passo 4: Instalar o FlexRIC

Execute os passos abaixo.

```bash
cd ~

git clone https://gitlab.eurecom.fr/mosaic5g/flexric.git

cd PATH_TO/flexric

git checkout v1.0.0

mkdir build

cd build

cmake ..

make

sudo make install
```

# Passo 5: Clonar o repositório do projeto OAI 5G RAN

Execute os comandos abaixo.

```bash
cd ~

git clone https://gitlab.eurecom.fr/oai/openairinterface5g.git oai

cd PATH_TO/oai

git checkout 2022.41
```

# Passo 6: Aplicar o patch do FlexRIC

Para habilitar o uso do FlexRIC com o OAI é necessário aplicar um patch no repositório clonado no passo anterior. Para isso, execute os passos abaixo.

```bash
cd PATH_TO/oai

git am PATH_TO/flexric/multiRAT/oai/oai.patch --whitespace=nowarn
```

# Passo 7: Executar o script de instalação

Para instalar a gNodeB junto com o FlexRIC, execute os comandos abaixo.

```bash
cd PATH_TO/oai

source oaienv

cd cmake_targets

sudo ./build_oai -I -w USRP -i

sudo ./build_oai --gNB -c -C -w USRP --ninja --build-lib all
```
